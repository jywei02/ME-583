```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-4, 4, 200)
x2 = np.linspace(-4, 4, 200)
X1, X2 = np.meshgrid(x1, x2)

V = 0.5 * (X1**2 + X2**2)

plt.figure()
contours = plt.contour(X1, X2, V, levels=[0.1, 0.5, 1, 2, 3, 5])
plt.clabel(contours, inline=True)

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def system(t, x):
    x1, x2 = x
    dx1 = -x1 + x2**2
    dx2 = -x2 - 4*x1*x2
    return [dx1, dx2]

t_span = (0, 10)
t_eval = np.linspace(t_span[0], t_span[1], 500)
initial_conditions = [(-2, -2), (-2, 2), (2, 2)]

x1 = np.linspace(-4, 4, 200)
x2 = np.linspace(-4, 4, 200)
X1, X2 = np.meshgrid(x1, x2)
V = 0.5 * (X1**2 + X2**2)

contours = plt.contour(X1, X2, V, levels=[0.1, 0.5, 1, 2, 3, 5])
plt.clabel(contours)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)

for x0 in initial_conditions:
    sol = solve_ivp(system, t_span, x0, t_eval=t_eval)
    plt.plot(sol.y[0], sol.y[1], label=f"x0 = {x0}")
    plt.plot(x0[0], x0[1], 'ro')
    plt.plot(sol.y[0][-1], sol.y[1][-1], 'kx')

plt.legend()
plt.show()
```
